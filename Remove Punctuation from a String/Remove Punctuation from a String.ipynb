# Remove Punctuation from a String

# Code

```python
string=input("Enter the string:")
punchuation=['.',',','!','?','/']
for n in range(0,len(string)):
    if string[n] in punchuation:
        n+=1 
    else:
        print(string[n],end='')
        n+=1
```
