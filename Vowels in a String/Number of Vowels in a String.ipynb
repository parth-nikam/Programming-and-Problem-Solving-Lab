# Number of Vowels in a String

# Code

```python
str=input('Enter the string:')
count=0
list1=['a','e','i','o','u','A','E','I','O','U']
for ch in str:
    if ch in list1:
        count=count+1
        
else:
    print('Number of vowels is:',count)
```
