# Reverse string Using user defined function

# Code

```python
string=input("Enter the string:") 
def reverse():
    print(string[::-1])
reverse()
```

# Algorithm


1. START.
2. Define function reverse( ).
3. Through negative slicing print the string. 
4. Call reverse( ).
5. STOP.
