```python
class PN_MOTORS:
    
    def __init__(self,model,color,ex_showroom,on_road):
        self.model=model
        self.color=color
        self.ex_showroom=ex_showroom
        self.on_road=on_road
    
    def geti(self):
        self.model=input("ENTER MODEL NAME:")
        self.color=input("ENTER COLOR:")
        self.ex_showroom=input("ENTER EX-SHOWROOM PRICE:")
        self.on_road=input("ENTER ON ROAD PRICE:")
    
    def showi(self):
        print("\n\nDETAILS ARE-\n")
        print("Model:",self.model)
        print("Color:",self.color)
        print("Ex-showroom price:",self.ex_showroom)
        print("On road price:",self.on_road)

car1=PN_MOTORS('abc','xyz',0,0)
car1.geti()
car1.showi()
print("STOP")
```
